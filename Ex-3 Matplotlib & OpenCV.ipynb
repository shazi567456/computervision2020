# Ex-5 Image Resize using Matplotlib & OpenCV

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('C:/Users/MY LENOVO/Desktop/butterfly.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

```python
new = cv2.resize(img2,(320,520))
```

```python
plt.imshow(new)
```

```python
new.shape
```

```python
new = cv2.resize(img2,(0,0),img2,w_change,h_change)
```
