# Ex-5 Creating Shapes using OpenCV & Matplotlib

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
cv2.rectangle(img,pt1=(384,10),pt2=(500,150),color=(0,255,0), thickness=9)
plt.imshow(img)
```

```python
cv2.rectangle(img,pt1=(200,200),pt2=(300,300),color=(0,0,255), thickness=9)
plt.imshow(img)
```

```python
cv2.circle(img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(img)
```

```python
cv2.circle(img,center=(100,100),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(img)
```

```python
cv2.line(img,pt1=(0,0),pt2=(511,511),color=(180,50,60),thickness=6)
plt.imshow(img)
```
